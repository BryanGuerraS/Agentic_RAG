```python
from langchain_cohere import CohereEmbeddings
from langchain_chroma import Chroma

# Inicialización del Chroma Local
chroma_local = Chroma(
    collection_name="documentos", 
    embedding_function=CohereEmbeddings(model="embed-multilingual-v2.0"), 
    persist_directory="chroma_db/preprocessed"
) 

# Mensaje de Inicio de Prueba
print("Iniciando prueba de carga del Vector DB...")

# Imprimir la instancia de Chroma
print(chroma_local)

# Imprimir el número de documentos en la base de datos
num_docs = chroma_local._collection.count()
print(f"Número de documentos en la base de datos: {num_docs}")

# Búsqueda de similitud para el archivo documento.docx
query = "Quien es Zara?"
results = chroma_local.similarity_search(query, k=5)

print("Metadatos de los documentos cargados:")

# Imprimir los metadatos de los documentos cargados
for result in results:
    print(result.metadata)

print("Prueba completada.")
```
